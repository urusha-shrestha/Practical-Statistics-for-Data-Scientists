# **Chapter 4 : Regression and Prediction**

Supervised Learning: the process of training a model on data where the outcome is known

### Simple Linear Regression

* provides a model of the relationship between the magnitude of one variable and that of a scond
* correlation measures the strength of an association between two variables while regression quantifies the nature of the relationship
* estimates how much Y will change when X changes by a certain amount
* tries to find the 'best' line to predict the response

Regression equation

 >  Y=b0+b1X

Y = response, the variable we are trying to predict, dependent variable, target, outcome

X = the variable used to predict the response, independent variable, feature, attribute, predictor

b0 = the intercept of the regression line, the predicted Y when X=0

b1 = regression coefficient, the slope of the line, weights, parameter estimates


Fitted values: the estimates obtained from the regression line, predicted values

Residuals: the difference between the observed values and the fitted values

Least squares: the method of fitting a regression by minimizing the sum of squared residuals

###Multiple Linear Regression

* when there are multiple predictors, the regression equation will be

> Y = b0 + b1X1 + b2X2 +...+ bpXp + e


Root mean squared error (RMSE): the square root of the average squared error of the regression

Residual standard error (RSE): the same as the RMSE, but adjusted for degrees of freedom

R-squared (coefficient of determination): the proportion of variance explained by the model, from 0 to 1

Weighted regression: regression with the records having different weights

###Cross Validation

* a statistical method used to evaluate the performance of a model by dividing the original sample into a training set, test set and a holdout set/ validation set.

K-Fold Cross Validation:

* 1/k of the data is set aside as a holdout sample
* the model is trained on the remaining data
* the model is applied to the holdout sample and necessary metrics are calculated
* the holdout sample is restored with the data and another 1/k data is set aside
* repeat the steps untill each record has been used in the holdout potion


###Model Selection and Stepwise Regression

* including additional variables always reduces RMSE and increases Rsquare for the training data

Subset Regression:

* search through all possible models
* computationally expensive and not feasible for problems with large data

Stepwise Regression:

* Backward elimination: Start with a full model and successively drop variables the don't contribute meaningfully

* Forward selection: Start with a constant model with no predictors and successively add variables

Penalized Regression:

* the model-fitting equation consists of a constraint that penalizes the model for too many variables
* applies penalty by reducing coefficients
* ridge regression, lasso regression


###Prediction using Regression

Prediction interval: an uncertainty interval around an individual predicted value

Extrapolation: extension of a model beyond the range of the data used to fit it

Confidence intervals:

* quantify uncertainty around regression coefficients
* uncertainty around a mean or other statistic calculated from multiple values

Prediction intervals:

* quantify uncertainty in individual predictions
* uncertainty around a single value



### Factor (Categorical) Variables in Regression

* regression required numerical inputs, so factor variables must be encoded

Dummy variables: binary 0-1 variables derived by recording factor data

Reference coding (treatment coding): one level of factor is used as a reference and other factors are compared to that level

One hot encoder:

* all factor levels are retained
* not appropriate for multiple linear regression

Deviation coding (sum contrats): a type of coding that compares each level against the overall mean as opposed to the reference level  



### Interpreting the Regression Equation

Multicollinearity:

* a situation in which two or more predictor variables are highly correlated with each other
* when one predictor variable can be expressed as a linear combination of others
* can cause numerical instability in fitting the regression equation

Confounding variables: an important predictor, that when removed leads to false relationships in a regression equation

Note: with correlated variables, the problem is of commission (including variables having similar predictive relationship with the response) and with condounding variables, the problem is of omission (an important variable being removed from the equation)

Main effects: the relationship between a predictor and the outcome variable, independent of other variables

Interactions: an interdependent relationship between two or more predictors and the response

### Regression Diagnostics

Standardized residuals: residuals divided by the standard error of the residuals

Influential value: a value whose presence or absence makes a big difference in the regression equation

Leverage (hat-value): the degree of influence that a single record has on a regression equation

Heteroskedasticity:

* when some ranges of the outcome experience residuals with higher variance
* lack of constant residual variance across the range of predicted values
* when errors are greater for some portions of the range than for others

Partial residual plots: a diagnostic plot to illuminate the relationship between the outcome and a single predictor

Note: normally distributed errors are a sign that the model is complete, non normally distributed errors indicate that the model may be missing something

### Polynomial and Spline Regression

* the relationship between the response and a predictor variable is not necessarily linear

Polynomial regression:

* involves including polynomial terms in a regression equation
* the relationship between the independent variable x and the dependent variable y is modeled as an nth degree polynomial

image.png

* only captures a certain amount of curvature in a non linear relationship, and adding higher-order terms often leads to undesirable wiggliness in the regression equation


Spline regression:

* an alternative to polynomial regression
* involves fitting a smooth curve with a series of polynomial segments
* provides a way to smoothly interpolate between fixed points
* a series of piecewise continuous polynomials

Knots: values that separate spline segments

Generalized additive models (GAM):

* spline models with automated selection of knots
